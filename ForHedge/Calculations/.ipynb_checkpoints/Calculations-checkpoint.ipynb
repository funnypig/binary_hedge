## Is Forex positions hedging with binary options profitable?

Assuming leverage is 100

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
LOT = 10**5

def forex_profit(start_price, current_price, lots, direction):
    '''
        Calculates forex profit or loss.
        Direction: 1 for LONG position and -1 for SHORT position
    '''
    
    return (current_price-start_price) * LOT * lots * direction

def hl_option_payout(barrier, cur_price, stake, direction):
    '''
        Calculates Higher/Lower option payout
    '''
    payout_coef = 1.85
    
    if cur_price < barrier:
        if direction == 1:
            return 0
        else:
            return round(stake * payout_coef, 2)
    else:
        if direction == -1:
            return 0
        else:
            return round(stake * payout_coef, 2)
        
```

### Imagine you have opened LONG position at 1.11
### Let's look at your profit depending on price

```python

start_price = 1.11
price = np.linspace(start_price - 300/LOT, start_price + 300/LOT, 600)

forex_profit = np.vectorize(forex_profit)
hl_option_payout = np.vectorize(hl_option_payout)

stake = 5

FP = forex_profit(start_price, price, 0.01, 1)
OPT = hl_option_payout(start_price, price, stake, -1)
```

```python
plt.plot(price, FP, label='FOREX')
plt.plot(price, OPT, label='OPTION')
plt.plot(price, FP+OPT-stake, label='HEDGE')

plt.title('Profit. Stake = 5')
plt.legend()
plt.show()
```

### Modeling:
#### Start price is fixed. Model 100 positions

```python

stake = 1
start_price = 1.11
price = np.linspace(start_price - 300/LOT, start_price + 300/LOT, 600)

spread = 0.0003
prices = np.random.uniform(start_price - 300/LOT, start_price + 300/LOT, size=100)

balance_forex = [100]
balance_option = [100]
balance_hedge = [100]

for i in range(100):
    
    p = start_price + 200/LOT if np.random.randint(2) else start_price - 200/LOT
    
    bf = balance_forex[-1]
    bo = balance_option[-1]
    bh = balance_hedge[-1]
    
    bf += forex_profit(start_price, p, 0.01, 1) - spread
    bo = bo - stake + hl_option_payout(start_price, p, stake, -1)
    bh = bh - stake - spread + \
        forex_profit(start_price, p, 0.01, 1) + hl_option_payout(start_price, p, stake, -1)
    
    balance_forex.append(bf)
    balance_option.append(bo)
    balance_hedge.append(bh)
    
plt.plot(np.arange(len(prices)+1), balance_forex, label='Forex')
plt.plot(np.arange(len(prices)+1), balance_option, label='Option')
plt.plot(np.arange(len(prices)+1), balance_hedge, label='Hedge')
plt.legend()
plt.show()
```

### As you can see the profit is lower if your forex predictions are right.

### But! The balance is much higher after loosing few times in a row!
